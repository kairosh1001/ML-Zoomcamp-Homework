Vector-vector multiplication

```python
def vector_vector_multiplication(u, v):
    assert u.shape[0] == v.shape[0]
    n = u.shape[0]
    result = 0.0

    for i in range(n):
        result += u[i] * v[i]
    return result
```

```python
u = np.array([1, 2, 3])
```

```python
v = np.array([4, 5, 6, 7])
```

```python
vector_vector_multiplication(u, v)
```

```python
u.dot(v)
```

Matrix-vector multiplication = Dot product b/w every matrix row and vector = m*n * n*1 = m*1
Row number from matrix, column number from vector (1)

```python
U = np.array([[2, 4, 5, 6], 
              [1, 2, 1, 2], 
              [3, 1, 2, 1]])
```

```python
U.shape
```

```python
def matrix_v_multiplication(U, v):
    assert U.shape[1] == v.shape[0]

    num_rows = U.shape[0]

    result = np.zeros(num_rows)

    for i in range(num_rows):
        result[i] = vector_vector_multiplication(U[i], v)
        
    return result
```

```python
matrix_v_multiplication(U, v)
```

```python
U.dot(v)
```
