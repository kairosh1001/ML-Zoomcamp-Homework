```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
data = 'https://raw.githubusercontent.com/alexeygrigorev/mlbookcamp-code/master/chapter-03-churn-prediction/WA_Fn-UseC_-Telco-Customer-Churn.csv'
```

```python
df = pd.read_csv(data)
df.head()
```

```python
df.head().T #shows columns as rows, convenient if many columns
```

```python
df.dtypes
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')

categorical_columns = list(df.dtypes[df.dtypes == 'str'].index)

for c in categorical_columns:
    df[c] = df[c].str.lower().str.replace(' ', '_')

categorical_columns
```

```python
df.totalcharges = pd.to_numeric(df.totalcharges, errors = 'coerce') #error-handling parameter to convert unconvertible data into missing values
df.totalcharges = df.totalcharges.fillna(0)
df.loc[df.totalcharges == 0, 'totalcharges']
```

```python
df.churn = (df.churn == 'yes').astype(int)
df.churn
```

```python
df_full_train, df_test = train_test_split(df, test_size = 0.2, random_state = 1)
```

```python
len(df_full_train), len(df_test)
```

```python
df_train, df_val = train_test_split(df_full_train, test_size = 0.25, random_state = 1)
```

```python
len(df_train), len(df_test), len(df_val)
```

```python
df_train = df_train.reset_index(drop=True)
df_val = df_val.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)
```

```python
y_train = df_train.churn.values
y_test = df_test.churn.values
y_val = df_val.churn.values
y_train
```

```python
df_train
```

```python
del df_train['churn']
del df_val['churn']
del df_test['churn']
df_train
```

Feature importance: churn rate and risk ratio

```python
churn_female = df_full_train[df_full_train.gender == 'female'].churn.mean()
churn_male = df_full_train[df_full_train.gender == 'male'].churn.mean()
churn_female, churn_male
```

```python
global_churn = df_full_train.churn.mean()
global_churn
```

```python
df_group = df_full_train.groupby('gender').churn.agg(['mean', 'count']) #.agg used to apply one or more operations across a specified axis (rows or columns)
df_group['diff'] = df_group['mean'] - global_churn
df_group['risk'] = df_group['mean'] / global_churn
df_group
```

```python
from IPython.display import display
for c in categorical:
    df_group = df_full_train.groupby(c).churn.agg(['mean', 'count']) 
    df_group['diff'] = df_group['mean'] - global_churn
    df_group['risk'] = df_group['mean'] / global_churn
    display(df_group)
```

One-hot encoding (Scikit-learn)

1 - activated cells (hot), 0 -cold cells

```python
from sklearn.feature_extraction import DictVectorizer #one way of one-hot encoding
```

```python
train_dicts = df_train[categorical + numerical].to_dict(orient = 'records')
```

```python
dv = DictVectorizer(sparse = False) #new instance of the class; not using sparse matrix
```

```python
X_train = dv.fit_transform(train_dicts)
```

```python
dv.get_feature_names_out()
```

```python
val_dicts = df_val[categorical + numerical].to_dict(orient = 'records')
```

```python
X_val = dv.transform(val_dicts)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

Using the model

```python
dicts_full_train = df_full_train[categorical + numerical].to_dict(orient = 'records')
dicts_full_train[:3]
```

```python
dv = DictVectorizer()
X_full_train = dv.fit_transform(dicts_full_train)
y_full_train = df_full_train.churn.values
model = LogisticRegression()
model.fit(X_full_train, y_full_train)
```

```python
dicts_test = df_test[categorical + numerical].to_dict(orient = 'records')
X_test = dv.transform(dicts_test)
y_pred = model.predict_proba(X_test)[:, 1] #model.predict_proba outputs a 2D array where each row contains two probabilities. [:, 1] shows rows of the positive class (will churn)
churn_decision_final = (y_pred >= 0.5)
```

```python
(churn_decision_final == y_test).mean()
```
